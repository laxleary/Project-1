```python
#Import dependencies
import pandas as pd

#Import csv data and use Pandas to create a Data Frame
df = pd.read_csv("../Resources/Original_CVD.csv")
df
```

```python
#Rename Columns to be more natural 
df = df.rename(columns = {
    'General_Health': 'General Health',
    'Heart_Disease': 'Heart Disease',
    'Skin_Cancer': 'Skin Cancer',
    'Other_Cancer': 'Other Cancer',
    'Age_Category': 'Age Category',
    'Height_(cm)': 'Height (cm)',
    'Weight_(kg)':'Weight (kg)', 
    'Smoking_History':'Smoking History',
    'Alcohol_Consumption':'Alcohol (Servings/Month)', 
    'Fruit_Consumption':'Fruit (Servings/Month)',
    'Green_Vegetables_Consumption': 'Green Vegetables (Servings/Month)', 
    'FriedPotato_Consumption': 'Fried Potato (Servings/Month)'
})

#Check for null rows and drop any:
df.dropna()

#Checked for odd values using .unique() 
#Identified desirable changes in the columns edited below

#Change the Last Checkup Column to be quickly readable
df["Checkup"] = df["Checkup"].replace("Within the past year", "<1" )
df["Checkup"] = df["Checkup"].replace("Within the past 2 years", "<2" )
df["Checkup"] = df["Checkup"].replace("Within the past 5 years", "<5" )
df["Checkup"] = df["Checkup"].replace("5 or more years ago", "5+" )

#Change the Diabetes Column to have simpler answers
df['Diabetes']=df['Diabetes'].replace({"No":"None", \
                                       "No, pre-diabetes or borderline diabetes":"Pre-diabetes or Borderline", \
                                        "Yes, but female told only during pregnancy":"Pregnacy-Induced"})

```

```python
#Export the Data frame to a csv for use in data analysis
df.to_csv("../Resources/Cleaned_CVD.csv", index=False)
```
